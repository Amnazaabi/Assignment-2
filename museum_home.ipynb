```python
# Import necessary classes
from Location import Location
from artwork import Artwork
from exhibition import Exhibition
from Visitor import Visitor
from Ticket import Ticket
from Event import Event

# Create an instance of Location
location = Location("Main Gallery")

# Create an instance of Artwork
artwork = Artwork("Mona Lisa", "Leonardo da Vinci", "c. 1503–1506", "Renaissance masterpiece", location)

# Create an instance of Exhibition
exhibition = Exhibition(location, "2 weeks")

# Create an instance of Visitor
visitor = Visitor(1, "Amna", 21)

# Create an instance of Ticket
ticket = Ticket("Art Exhibition", 50.00)

# Create an instance of Event
event = Event("Art Exhibition", location, "2 hours", 50.00)

# Print information of each instance
print("Location:", location.get_name())
print()

print("\n------------Artwork----------------")
print("Title:", artwork._title)
print("Artist:", artwork._artist)
print("Date of Creation:", artwork._date_of_creation)
print("Historical Significance:", artwork._historical_significance)
print("Exhibition Location:", artwork._exhibition_location.get_name())
print()

print("\n------------Exhibition----------------")
print("Location:", exhibition._location.get_name())
print("Duration:", exhibition._duration)
print()

print("\n------------Visitor--------------")
print("ID:", visitor._id)
print("Name:", visitor._name)
print("Age:", visitor._age)
print()

print("\n------------Ticket---------------")
print("Event:", ticket._event)
print("Price:", ticket._price)
print()

print("\n------------Event-----------------")
print("Name:", event._name)
print("Location:", event._location.get_name())
print("Duration:", event._duration)
print("Ticket Price:", event._ticket_price)
```

```python
class Ticket:
    def __init__(self, event, price):
        """
        Initializes a Ticket object with the given event and price.

        Args:
            event (str): The event for which the ticket is purchased.
            price (float): The price of the ticket.
        """
        self._event = event
        self._price = price

    def get_event(self):
        """
        Gets the event for which the ticket is purchased.

        Returns:
            str: The event for which the ticket is purchased.
        """
        return self._event

    def set_event(self, event):
        """
        Sets the event for which the ticket is purchased.

        Args:
            event (str): The new event for the ticket.
        """
        self._event = event

    def get_price(self):
        """
        Gets the price of the ticket.

        Returns:
            float: The price of the ticket.
        """
        return self._price

    def set_price(self, price):
        """
        Sets the price of the ticket.

        Args:
            price (float): The new price for the ticket.
        """
        self._price = price

    def apply_discount(self, category):
        """
        Applies a discount to the ticket price based on the provided category.

        Args:
            category (str): The category of the ticket buyer.

        Discounts:
            - For 'student' category: 20% discount
            - For 'senior' category: 10% discount
            - For 'group' category: 50% discount
        """
        if category == "student":
            self._price *= 0.8  # Applying 20% discount for students
        elif category == "senior":
            self._price *= 0.9  # Applying 10% discount for seniors
        elif category == "group":
            self._price *= 0.5  # Applying 50% discount for groups
      
```

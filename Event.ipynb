```python
from Location import Location  # Import the Location class

class Event:
    def __init__(self, name, location, duration, ticket_price):
        """
        Initializes an Event object with the given attributes.

        Args:
            name (str): The name of the event.
            location (Location): The location of the event.
            duration (str): The duration of the event.
            ticket_price (float): The price of tickets for the event.
        """
        self._name = name
        self._location = location
        self._duration = duration
        self._ticket_price = ticket_price

    def get_name(self):
        """
        Gets the name of the event.

        Returns:
            str: The name of the event.
        """
        return self._name

    def set_name(self, name):
        """
        Sets the name of the event.

        Args:
            name (str): The new name for the event.
        """
        self._name = name

    def get_location(self):
        """
        Gets the location of the event.

        Returns:
            Location: The location of the event.
        """
        return self._location

    def set_location(self, location):
        """
        Sets the location of the event.

        Args:
            location (Location): The new location for the event.
        """
        self._location = location

    def get_duration(self):
        """
        Gets the duration of the event.

        Returns:
            str: The duration of the event.
        """
        return self._duration

    def set_duration(self, duration):
        """
        Sets the duration of the event.

        Args:
            duration (str): The new duration for the event.
        """
        self._duration = duration

    def get_ticket_price(self):
        """
        Gets the ticket price of the event.

        Returns:
            float: The ticket price of the event.
        """
        return self._ticket_price

    def set_ticket_price(self, ticket_price):
        """
        Sets the ticket price of the event.

        Args:
            ticket_price (float): The new ticket price for the event.
        """
        self._ticket_price = ticket_price

    def get_details(self):
        """
        Gets the details of the event.

        Returns:
            str: A string containing details of the event.
        """
        return f"Name: {self._name}, Location: {self._location.get_name()}, Duration: {self._duration}, Ticket Price: {self._ticket_price}"
```
